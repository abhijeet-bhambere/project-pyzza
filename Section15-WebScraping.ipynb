## Downloading multiple pages 
Bringin' in the `for` loops. Gt the title of every book on [books.toscrape.com]("https://books.toscrape.com/") with 4-stars

```python
import requests
import bs4
```

```python
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
```

_So this sets up the basic funcs required to parse each webpage & get the titles with 4-star ratings_<br>
<br>
_Now we need to set these funcs into a FOR loop to parse each of the 50 webpages & find 4-star titles_


Check if the book being iterated on is a 4-star book.<br>
Do this by trying FIRST intermediate ResultSet all elements with star-rating.Four class.<br>
The element will get added to this intermediate ResultSet if there is a `.star-rating.Four` class for given book.<br>
If the ResultSet contain a non-zero length -- meaning there is some 4-star book, only then go check for the title.

```python
four_star_titles = []

for n in range(1,51):
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup = bs4.BeautifulSoup(res.text,'lxml')
    books = soup.select('.product_pod')
    
    # For ever book in books ResultSet, try to look for elements containing star-rating.Four class
    # If it passes above check , only thne Add the book into the list  
    for book in books:
        ## Alternate method: to look for 'star-rating Four' string in book string
        # if 'star-rating Four' in str(book):
        if len(book.select('.star-rating.Four'))!=0:
            book_title = book.select('a')[1]['title']
            four_star_titles.append(book_title)
```

```python
four_star_titles
```

_Here's the list of all 4-star book titles from the bookscrape site_

```python
# Printing out the results:
for every in four_star_titles:
    print(every)
```
